```python
import numpy as np
import matplotlib.pyplot as plt

from torchvision.datasets import FashionMNIST
```

### Load data

```python
dataset = FashionMNIST('../data', download=True, train=True)
```

```python
CLASSES = [
    'T-shirt/top', 'Trousers', 'Pullover', 'Dress', 'Coat', 'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot'
]
```

### General data information

```python
sample = dataset[0]

img = np.array(sample[0])
target = sample[1]
```

```python
print(f'shape: {img.shape}\nmin - max: {img.min()}-{img.max()}\ndtype: {img.dtype}')
```

```python
for l in range(10):
    n = sum([dataset[i][1] == l for i in range(len(dataset))])
    print(f"{l} - {n / len(dataset)}%")
```

### Data vizualization

```python
plt.clf()

fig, axs = plt.subplots(4, 6, figsize=(18, 12))

for i in range(24):
    img, target = dataset[i]
    axs.flat[i].imshow(img, cmap='gray')
    axs.flat[i].set_title(f'{target}: {CLASSES[target]}')
    axs.flat[i].set_axis_off()
plt.show()
```
